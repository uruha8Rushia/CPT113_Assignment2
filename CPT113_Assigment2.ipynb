# Problem solved by using Object-Oriented Programming

<p align="justify">
There are some problems in the program that is successfully solved by using Object-Oriented Programming, including complexity of managing large number of songs in playlist and maintaining a scalable organization of the song collections. Through using Object-Oriented Programming, Jane able to manage her playlist more effectively even if a large number of songs is added into the music player.

1. Songs Management
- Problem: It is difficult for the user to manage a large number of songs in their song collection manually. When user wants to add, remove or edit any songs in a playlist with 1000 songs, it is unrealistic for them to handle the task manually as it wastes time and energy.
- Solution: A class of doubly linked list helps to handle the addition, removal or editing of songs efficiently. By using doubly linked list to manage the music playlist, each song can be represented as a node in the linked list, where every node is linked together following the name order. This allows various operations such as insertion, deletion and traversal on the playlist, ensuring the effective management of the large size playlist.

2. Scalable Organization of Playlist
- Problem: User should be able to perform future enhancements on the program easily without affecting the main structure of the program. Besides, the program should be flexible to handle the increasing number of songs in the playlist.
- Solution: Different classes is used to encapsulate different data in the program. For instance, the song details are encapsulated in the 'Media' and 'Song' classes, while the collection of songs is managed in the 'MusicPlayer' class as nodes. This makes the program easy to maintain and scale for future use.

3. Play Features
- Problem: The program have to track the previous and upcoming songs for the song that is currently played throughout the program, so that the previous or next can be played if respective action is chosen by the user.
- Solution: A doubly linked list data structure is used, where each node in the list contains references to both the previous node and the next node. This allows easy navigation forwards and backwards through the playlist, which is useful for the play features.

# Parent Class Media


The Media class is a templated class that represents media items such as songs, videos, or any other type of media. It is the parent class of Song class and contains three private member variables: name, artist, and genre, which store the name of the media item, the name of the artist, and the genre of the media item respectively.

It encapsulates the common attributes and behaviours of various media items, providing a flexible and reusable foundation for working with media in the program.

```python
%%writefile Media.h
#ifndef MEDIA_H
#define MEDIA_H
#include <iostream>
using namespace std;

template <class T>
class Media{
  private:
    T name; //name of the song
    T artist; //name of the artist
    T genre; //genre of the song
  public:
    Media() {} //default constructor
    Media(T, T, T); //overloaded constructor
    ~Media() {} //destructor

    //getter functions
    T getName();
    T getArtist();
    T getGenre();

    void setName(string); //setter function for name

};
#endif
```

```python
%%writefile Media.cpp
#include "Media.h"

//Explicit instantiation of Media class template with string
template class Media<string>;

//Getter for the name
template <class T>
T Media<T>::getName()
{
  return name;
}

//Getter for the artist
template <class T>
T Media<T>::getArtist()
{
  return artist;
}

//Getter for the genre
template <class T>
T Media<T>::getGenre()
{
  return genre;
}

//Overloading constructor to initialize media with name, artist and genre
template <class T>
Media<T> :: Media(T abc , T def, T chg)
{
  name = abc ;
  artist = def;
  genre = chg;
}

//Setter for the name
template <class T>
void Media<T>:: setName(string changes)
{
  name = changes;
}
```

# Class ListNode

The ListNode class is a templated class that serves as the building block of the doubly linked list. It represents a single node containing data of type 'T' and two pointers, which one points to the next node and the other points to the previous node.

The ListNode class is composed of a Song object, where each node in the doubly linked list encapsulates a 'Song' object in this scenario. It provides the basic functionality needed to create and manage the nodes in the doubly linked list. By using the ListNode class, different operations such as insertion, deletion and traversal can be performed efficiently.

```python
%%writefile ListNode.h
#ifndef LISTNODE_H
#define LISTNODE_H
#include "Song.h"
#include <iostream>
using namespace std;

template <class T>
class ListNode
{
  private:
    T value; //value of the node
    ListNode<T> *next; //pointer to the next node
    ListNode<T> *back; //pointer to the previous node
  public:
    ListNode(T abc); //constructor
    ~ListNode(); //destructor

    //setter functions
    void setNext(ListNode<T>*);
    void setBack(ListNode<T>*);
    void setValue(T);

    //getter functions
    ListNode<T>* getNext();
    ListNode<T>* getBack();
    T getValue();
};
#endif
```

```python
%%writefile ListNode.cpp
#include "ListNode.h"

//Explicit instantiation of ListNode struct template with Song class
template class ListNode< Song <string,double>>;

//constructor
template <class T>
ListNode<T> :: ListNode(T abc)
{
  value = abc;
  next = nullptr;
  back = nullptr;
}

//destructor
template <class T>
ListNode<T> :: ~ListNode(){}

//setter function for next pointer
template <class T>
void ListNode<T> :: setNext(ListNode<T> *ptr)
{
  next = ptr;
}

//setter function for back pointer
template <class T>
void ListNode<T> :: setBack(ListNode<T> *ptr)
{
  back = ptr;
}

//setter function for value of node
template <class T>
void ListNode<T> :: setValue(T abc)
{
  value = abc;
}

//getter function for next pointer
template <class T>
ListNode<T>* ListNode<T> :: getNext()
{
  return next;
}

//getter function for back pointer
template <class T>
ListNode<T>* ListNode<T> :: getBack()
{
  return back;
}

//getter function for value of node
template <class T>
T ListNode<T> :: getValue()
{
  return value;
}
```

# Class MusicPlayer

The MusicPlayer class is a templated class that provides functionalities to manage a collection of songs using a doubly linked list. It is composed of the ListNode class, where multiple ListNode objects are used to manage the collection of songs in the doubly linked list. It consists of pointers that point to the first node, last node and the current node respectively.

The MusicPlayer class is used as the main object declared in the main program. It provides a comprehensive interface for managing a music playlist, including various actions such as adding, removing and playing songs. Doubly linked list is implemented in the class for effective data management.

```python
%%writefile MusicPlayer.h
#ifndef MUSICPLAYER_H
#define MUSICPLAYER_H

#include <iomanip>
#include <unistd.h>
#include <fstream>
#include <cstdlib>
#include <ctime>
#include <string>
#include "Song.h"
#include "ListNode.h"

template <class T>
class MusicPlayer{
  private:
    ListNode<T> *head; //pointer to the first node
    ListNode<T> *tail; //pointer to the last node
    ListNode<T> *currentPtr; //pointer to the current node

  public:
    MusicPlayer(ifstream&); //overloading constructor
    MusicPlayer(); //default constructor
    ~MusicPlayer(); //destructor

    void addSong(T); //add song function
    void displaySongList(); //display song list function

    bool searchSong(string ); //search song function

    void removeSong(string); //remove song function
    void removeSong(int); //overloading function to remove song

    void renameSong(int,string); //rename song function
    void renameSong(string, string); //overloading function to rename song

    void play(); //play song function
    void play(int); //overloading function to play song
    void playPrevious(); //play previous song function
    void playNext(); //play next song function
    void playShuffle(); //play random song function

    void updateFile(); //update file function
};

#endif
```

```python
%%writefile MusicPlayer.cpp
#include "MusicPlayer.h"

//Explicit instantiation of MusicPlayer class template with Song class
template class MusicPlayer < Song <string,double>>;

template <class T>
void MusicPlayer<T>::updateFile()
{
  ofstream outFile; //open an output file
  outFile.open("songAlbum.txt");

  //check if the file is successfully opened
  if (!outFile.fail())
  {
    cout<<"Updating file........................\n";

    //pointer to traverse the list
    ListNode<T> *nodePtr;
    nodePtr = head;

    //traverse the list and write each song's details to the file
    while (nodePtr != nullptr)
    {
      outFile <<nodePtr->getValue().getName()<<","
              <<nodePtr->getValue().getArtist()<<","
              <<nodePtr->getValue().getGenre()<<","
              <<nodePtr->getValue().getTime()<<endl;
      nodePtr = nodePtr->getNext();
    }

    outFile.close(); //close the output file
    cout<<"Done !"<<endl;
  }
  else
  {
    //print error message if file couldn't be opened
    cout<<"File open error. Please make sure input file desired exist under the same folder."<<endl;
  }
}

template <class T>
void MusicPlayer<T>::playShuffle()
{
  ListNode <T> *nodePtr; //pointer to traverse the list
  int counter = 0; //counter to count the number of nodes
  srand(time(0)); //seed the random number henerator with current time

  nodePtr = head; //initialise nodePtr to the head of the list

  //traverse the list to count the number of nodes
  while (nodePtr != nullptr)
  {
    counter++;
    nodePtr = nodePtr->getNext();
  }

  if (counter==0)
  {
    cout<<"List is empty\n"; //check if the list is empty
  }
  else if (counter > 1)
  {
    int randomNum = rand()%counter + 1; //generate a random number between 1 and the number of nodes
    play(randomNum); //call the play function with generated random number
  }
}

template <class T>
void MusicPlayer<T>::renameSong(int selected,string changes)
{
  ListNode <T> *nodePtr; //pointer to traverse the list
  int counter = 1; //counter to track the position of nodes

  //check if the list is empty
  if (!head)
    cout <<"List is empty\n";

  nodePtr = head ; //initialise the pointer to the head of the list

  //traverse the list until reaching the desired position or the end of the list
  while (nodePtr != nullptr && selected != counter)
  {
    nodePtr = nodePtr->getNext();
    counter++;
  }

  //if the desired position is found
  if(nodePtr)
  {
    //rename the song based on user's input
    nodePtr->getValue().rename(changes);
    cout<<"Done !"<<endl;
  }
  else
  {
    //print error message if no song is found at the selected position
    cout<<"No such Song included.\n";
  }
}

template <class T>
void MusicPlayer<T>::renameSong(string selected, string changes)
{
  ListNode <T> *nodePtr; //pointer to traverse the list

  //check if the list is empty
  if (!head)
    cout <<"List is empty\n";

  nodePtr = head; //initialise the pointer to the head of the list

  //traverse the list until input name is found or end of the list is reached
  while (nodePtr != nullptr && nodePtr->getValue().getName() != selected)
  {
    nodePtr = nodePtr->getNext();
  }

  if(nodePtr)
  {
    //rename the song based on user's input
    nodePtr->getValue().rename(changes);
    cout<<"Done !"<<endl;
  }
  else
  {
    //print error message if no song is found at the selected position
    cout<<"No such Song included.\n";
  }
}

template <class T>
void MusicPlayer<T>::removeSong(int searchValue)
{
  ListNode<T> *nodePtr; //pointer to traverse the list
  int counter = 1; //counter to track the position of nodes

  //check if the list is empty
  if (!head)
  cout <<"List is empty\n";

  //check if the first node is the target
  if (searchValue == 1)
  {
    nodePtr=head;
    head=head->getNext(); //update head to the next node
    head->setBack(nullptr); //set the back pointer of the new head to nullptr
    delete nodePtr;
    cout<<"Done !"<<endl;
  }
  else
  {
    //initialize nodePtr to the head of list
    nodePtr = head;

    //skip all nodes until reaching the desired position or the end of the list
    while (nodePtr != nullptr && searchValue != counter)
    {
      nodePtr = nodePtr->getNext();
      counter++;
    }

    //if nodePtr is at the end of the list, update tail
    if (nodePtr == tail)
    {
      tail = nodePtr->getBack(); //update tail to previous node
      tail->setNext(nullptr); // set the next pointer of the new tail to nullptr
      delete nodePtr;
      cout<<"Done !"<<endl;
    }
    //if nodePtr is betweeen the beginning and the end of the list
    else if (nodePtr)
    {
      //link the previous node to the next node after nodePtr
      nodePtr->getBack()->setNext(nodePtr->getNext());
      nodePtr->getNext()->setBack(nodePtr->getBack());
      delete nodePtr;
      cout<<"Done !"<<endl;
    }
    else
    {
      cout<<"No such Song included.\n"; //Output message if no song found at the selected position
    }
  }
}

template <class T>
void MusicPlayer<T>::removeSong(string searchValue)
{
  ListNode<T> *nodePtr;//pointer to traverse the list

  //check if the list is empty
  if (!head)
  cout <<"List is empty\n";

  //check if the first node is the target
  if (head->getValue().getName() == searchValue)
  {
    nodePtr=head; //update nodePtr to the first node
    head=head->getNext(); //update head to the next node
    head->setBack(nullptr); //set the back pointer of the new head to nullptr
    delete nodePtr;
    cout<<"Done !"<<endl;
  }
  //check if the last node is the target
  else if (tail->getValue().getName() == searchValue)
  {
    nodePtr=tail; //update nodePtr to the last node
    tail=tail->getBack(); //update tail to previous node
    tail->setNext(nullptr); //set the next pointer of the new tail to nullptr
    delete nodePtr;
    cout<<"Done !"<<endl;
  }
  else
  {
    //initialize nodePtr to the head of list
    nodePtr = head;

    //skip all nodes until reaching the desired position or the end of the list
    while (nodePtr != nullptr && nodePtr->getValue().getName()!= searchValue)
    {
      nodePtr = nodePtr->getNext();
    }

    //if nodePtr is at the end of the list, update tail
    if (nodePtr)
    {
      //link the previous node to the next node after nodePtr
      nodePtr->getBack()->setNext(nodePtr->getNext());
      nodePtr->getNext()->setBack(nodePtr->getBack());
      delete nodePtr;
      cout<<"Done !"<<endl;
    }
    else
    {
      cout<<"No such Song included.\n"; //Output message if no song found at the selected position
    }
  }
}

template <class T>
void MusicPlayer<T>::playPrevious()
{
  //check if there's a current song playing
  if (currentPtr)
  {
    currentPtr=currentPtr->getBack(); //move currentPtr to the previous node
    //check if currentPtr is nullptr
    if(currentPtr != nullptr)
      currentPtr->getValue().displaySong(); //display the current song
    else
    {
      currentPtr = tail; //set currentPtr to the last node
      currentPtr->getValue().displaySong(); //display the last node
    }
  }
  else
  {
    cout<<"Please play a song before choosing previous song.\n";
  }
}

template <class T>
void MusicPlayer<T>::playNext()
{
  //check if there's a current song playing
  if (currentPtr)
  {
    currentPtr = currentPtr->getNext(); //move currentPtr to the next node
    //check if currentPtr is nullptr
    if(currentPtr != nullptr)
      currentPtr->getValue().displaySong(); //display the current song
    else
    {
      currentPtr = head; //set currentPtr to the first node
      currentPtr->getValue().displaySong(); //display the first node
    }
  }
  else
  {
    cout<<"Please play a song before choosing next song.\n";
  }

}

template <class T>
void MusicPlayer<T>::play(int selected)
{
  ListNode<T> *nodeptr; //pointer to traverse the list
  nodeptr = head; //initialise nodeptr to the head of the list

  //traverse the list until reaching the desired position or the end of the list
  for (int i =1; i!=selected && nodeptr !=nullptr ;i++)
  {
    nodeptr =nodeptr->getNext();
  }
  if (nodeptr)
  {
    nodeptr->getValue().displaySong(); //display the current song
    currentPtr = nodeptr; //update currentPtr to the nodeptr
  }
  else
    cout<<"Please enter valid input and try again."<<endl;

}

template <class T>
void MusicPlayer<T>::play()
{
  //check if there's a current song playing
  if (currentPtr)
    currentPtr->getValue().displaySong();
  else
    cout<<"Please select song to be play"<<endl;
}

template <class T>
bool MusicPlayer<T>::searchSong( string searching)
{
  ListNode<T> *nodePtr; //pointer to traverse the list
  searching = "\""+searching+"\"";

  //check if the list is empty
  if (!head)
  {
    cout <<"List is Empty\n";
    return 0;
  }

  //check if the first node is the target
  if (head->getValue().getName()== searching)
  {
    cout<<"\n Found at first  \n";
    head->getValue().displaySong(); //display the first node
    currentPtr = head; //update currentPtr to the first node
    return 1;
  }
  else
  {
    nodePtr = head; //initialize nodePtr to the head of list
    int counter = 1; //counter to track the position of nodes

    //skip all nodes until reaching the desired position or the end of the list
    while (nodePtr != nullptr && nodePtr->getValue().getName()!= searching)
    {
      nodePtr = nodePtr->getNext();
      counter++;
    }

    //if nodePtr is at the end of the list, update tail
    if (nodePtr)
    {
      cout<<"\n Found at Song nombor "<<counter<<"\n";
      nodePtr->getValue().displaySong(); //display the current node
      currentPtr = nodePtr; //update currentPtr to the nodePtr
      return 1;
    }
    else
    {
      cout<<"\nNot found\n";
      return 0;
    }

  }
}

//destructor
template <class T>
MusicPlayer<T>::~MusicPlayer()
{
  //pointer to traverse the list
  ListNode<T> *nodePtr;
  ListNode<T> *nextNode;

  //position nodePtr at the head of the list
  nodePtr = head;

  //while nodePtr is not at the end of the list
  while (nodePtr != nullptr)
  {
    //save a pointer to the next node.
    nextNode = nodePtr->getNext();

    //delete the current node.
    delete nodePtr;

    //position nodePtr at the next node.
    nodePtr = nextNode;
  }
  head =nullptr;
  tail = nullptr;
  currentPtr = nullptr;
}

//default constructor
template <class T>
MusicPlayer <T>::MusicPlayer()
{
  head = nullptr;
  tail = nullptr;
  currentPtr = nullptr;
}

//overloading constructor
template <class T>
MusicPlayer <T>::MusicPlayer(ifstream& inFile): head(nullptr), tail(nullptr),currentPtr(nullptr)
{

  //check if the file is successfully opened
  if (inFile.fail())
  {
    cout<<"File open error. Please make sure input file desired exist under the same folder."<<endl;
  }
  else
  {
    string tempName,tempArtist,tempGenre;
    double tempTime;

    //read the data from file to temporary variables line by line
    while(getline(inFile,tempName,','))
    {
      getline(inFile,tempArtist,',');
      getline(inFile,tempGenre,',');
      inFile>>tempTime;
      inFile.ignore();

      //create a temporary song using the read data
      T tempSong(tempName,tempArtist,tempGenre,tempTime);

      //add the temporary song to the music player playlist
      addSong(tempSong);
    }
  }
}

template <class T>
void MusicPlayer <T>::addSong(T newSong){
  ListNode<T> *newNode; //a new node
  ListNode<T> *nodePtr; //pointer to traverse the list
  string duplicateSongException; //exception message for duplicated songs

  //allocate a new node and store newValue there
  newNode = new ListNode<T>(newSong) ;

  //if there are no nodes in the list, make newNode the first node
  if (!head)  {
    head = newNode;
    tail = newNode;
  }
  else //otherwise, insert newNode
  {
    nodePtr = head;

    //check for duplicated song names and artists
    while(nodePtr !=nullptr )
    {
      if (nodePtr->getValue().getName() == newSong.getName() && nodePtr->getValue().getArtist() == newSong.getArtist())
      {
        duplicateSongException = "Song you wish to add has already exist with the same name and artist.\nPlease add a new Song instead of duplicating existing one.\n";
        throw duplicateSongException; //throw exception if a duplicated song is found
      }
      nodePtr =nodePtr -> getNext();
    }
    //position nodePtr at the head of list.
    nodePtr = head;

    //skip all nodes whose value is less than newValue
    while (nodePtr != nullptr && nodePtr->getValue().getName() < newSong.getName())  {
      nodePtr = nodePtr->getNext();
    }

    //if the new node is to be the 1st in the list, insert it before all other nodes.
    if (nodePtr == head)  {
      head = newNode;
      newNode->setNext(nodePtr);
      nodePtr->setBack(newNode);
    }
    //if the new node is to be the last in the list, insert it after all other nodes.
    else if(nodePtr == nullptr){
      newNode->setBack(tail);
      tail->setNext(newNode);
      tail = newNode;
    }

    else //otherwise insert after the allocated node.
    {
      nodePtr->getBack()->setNext(newNode);
      newNode->setBack(nodePtr->getBack());
      nodePtr->setBack(newNode);
      newNode->setNext(nodePtr);
    }
  }
}

template <class T>
void MusicPlayer<T> :: displaySongList()
{
  //header of the song list display
  cout << "-------------------------------------------------------------------------------------------------------------------------------------------------------------------------\n";
  cout << "|                                                                               Music Playlist                                                                          |\n";
  cout << "-------------------------------------------------------------------------------------------------------------------------------------------------------------------------\n";
  cout << "| No. | Title                                              | Artist                                             | Genre                          | Length               |\n";
  cout << "-------------------------------------------------------------------------------------------------------------------------------------------------------------------------\n";

  ListNode<T> *nodePtr; //pointer to traverse the list
  nodePtr = head; //initialise nodePtr as the first node
  int counter = 1; //counter to track the position of nodes

  //while nodePtr points to a node, traverse the list
  while (nodePtr)
  {
    //display the value of the current node
    cout << "| " << setw(3) << left << counter;
    cout << " | " << setw(50) << left << nodePtr->getValue().getName();
    cout << " | " << setw(50) << left << nodePtr->getValue().getArtist();
    cout << " | " << setw(30) << left << nodePtr->getValue().getGenre();
    cout << setprecision(2) << fixed << showpoint;
    cout << " | " << setw(20) << left << nodePtr->getValue().getTime();
    cout << " |\n";

    nodePtr = nodePtr->getNext(); //move nodePtr to the next node
    counter++;
  }
  cout << "-------------------------------------------------------------------------------------------------------------------------------------------------------------------------\n";
  updateFile(); //update the input file after displaying the list
}
```

```python
%%writefile mainFunction.cpp
#include "Song.h"
#include "Media.h"
#include "ListNode.h"
#include "MusicPlayer.h"
#include <unistd.h>
#include <string>
#include <iostream>
#include <fstream>
using namespace std;

//global function for displaying main menu
void showMainMenu() {
    cout << "\n-------------------------------------------------------------------------------------------------------" << endl;
    cout << "-------------------------------------------------------------------------------------------------------" << endl;
    cout << "|                                        <  Main Menu >                                                |" << endl;
    cout << "-------------------------------------------------------------------------------------------------------" << endl;
    cout << "|         (1) Play Songs                                                                               |" << endl;
    cout << "|         (2) Play Previous                                                                            |" << endl;
    cout << "|         (3) Play Next                                                                                |" << endl;
    cout << "|         (4) Shuffle Play                                                                             |" << endl;
    cout << "|         (5) Search Song                                                                              |" << endl;
    cout << "|         (6) Edit Options                                                                             |" << endl;
    cout << "|         (99) Exit Program                                                                            |" << endl;
    cout << "-------------------------------------------------------------------------------------------------------" << endl;
    cout << "Enter your choice : ";
}

//global function for displaying sub menu of editing songs
void showOperationMenu() {
    cout << "-------------------------------------------------------------------------------------------------------" << endl;
    cout << "|                                       < Editor Menu >                                                |" << endl;
    cout << "-------------------------------------------------------------------------------------------------------" << endl;
    cout << "|         (1) Add a New Song                                                                           |" << endl;
    cout << "|         (2) Rename a Song                                                                            |" << endl;
    cout << "|         (3) Remove a Song                                                                            |" << endl;
    cout << "|         (4) Quit to Main Menu                                                                        |" << endl;
    cout << "|         (99) Exit Program                                                                            |" << endl;
    cout << "-------------------------------------------------------------------------------------------------------" << endl;
    cout << "Enter your choice: ";
}

int main ()
{
  //open the file containing song data
  ifstream inFile;
  inFile.open("songAlbum.txt");

  //create a MusicPlayer object using the file containing song data
  MusicPlayer < Song <string,double>>  Obj (inFile);

  //variables declaration
  int choice,songChoice, convertedRenameInt, convertedDeleteInt;
  double addedTime;
  string userSearching, selectedtoChange, changeTo, removedSelected;
  string addedName, addedArtist, addedGenre;
  Song<string,double> songExp;

  //main menu loop
  do
  {
    //display the song list and main menu options
    Obj.displaySongList();
    showMainMenu();
    cin>>choice;

    switch (choice)
    {
      case 1 : //play a specific song
        cout<<"Enter index of the song : ";
        cin>>songChoice;
        sleep(0.75);
        cout<<"Playing..............."<<endl;
        sleep(2);
        Obj.play(songChoice);
        sleep(4);
        break;
      case 2 : //play the previous song
        cout<<"Playing..............."<<endl;
        sleep(2);
        Obj.playPrevious();
        sleep(6);
        break;
      case 3 : //play the next song
        cout<<"Playing..............."<<endl;
        sleep(2);
        Obj.playNext();
        sleep(6);
        break;
      case 4 : //play a random song
        cout<<"Shuffling..............."<<endl;
        sleep(1.0);
        cout<<"Playing..............."<<endl;
        sleep(1.5);
        Obj.playShuffle();
        sleep(5);
        break;
      case 5 : //search for a song by name
        cout<<"Enter song's name to be found : ";
        cin.ignore();
        getline(cin,userSearching);
        sleep(0.75);
        cout<<"Searching............"<<endl;
        sleep(2);

        if (Obj.searchSong(userSearching))
        {
          //ask if the user wants to play the found song
          cout<<"\nDo you want to play it now ? (Yes/No) -> ";
          cin>>userSearching;

          if (userSearching == "Yes")
          {
            sleep(0.75);
            cout<<"Playing..............."<<endl;
            sleep(2);
            Obj.play();
            sleep(5);
          }
          else
          {
            cout<<"Exiting.........\n";
            sleep(1.5);
          }

        }
        else
        {
          cout<<"\nExiting.........\n";
          sleep(1.5);
        }
        break;
      case 6 : //open sub-menu for add, remove and rename song operations
        showOperationMenu();
        cin>>choice;

        switch(choice)
        {
          case 1: //add a new song
            //get user's input for the new song details
            cout<<"To add a new song, please provide : \n";
            cout<<"Song name : ";
            cin.ignore();
            getline(cin,addedName);
            addedName = "\""+addedName+"\"";
            cout<<"Song made by artist -> ";
            //cin.ignore();
            getline(cin,addedArtist);
            addedArtist = " "+ addedArtist;
            cout<<"Song genre : ";
            //cin.ignore();
            getline(cin,addedGenre);
            cout<<"Song time length : ";
            cin>>addedTime;

            cout<<"Adding new song......................\n";
            sleep(2);

            //create a new song object using the user's input
            songExp = Song<string,double> (addedName,addedArtist,addedGenre,addedTime);
            try {
              Obj.addSong(songExp);
              cout<<"Done!\n";
            }
            catch (string& duplicateSongException)
            {
              cout<<duplicateSongException;
            }

            sleep(1.5);
            break;

          case 2: //rename a song
            cout<<"Which of the songs you wish to change?  ->";
            cin.ignore();
            getline(cin,selectedtoChange);
            cout<<"Changed to ? -> ";
            getline(cin,changeTo);
            changeTo = "\""+changeTo+"\"";

            cout<<"Renaming................"<<endl;
            sleep(1.5);

            //convert input to integer if possible
            try {
              convertedRenameInt = stoi(selectedtoChange);
              Obj.renameSong(convertedRenameInt,changeTo);
            }
            catch (const invalid_argument&)
            {
              selectedtoChange = "\""+selectedtoChange+"\"";
              Obj.renameSong(selectedtoChange,changeTo);
            }

            sleep(3);
            break;

          case 3: //remove a song
            cout<<"Which of the songs you wish to remove? ->";
            cin.ignore();
            getline(cin,removedSelected);

            cout<<"Removing.....................\n";
            sleep(1.5);

            //convert input to integer if possible
            try {
              convertedDeleteInt = stoi (removedSelected);
              Obj.removeSong(convertedDeleteInt);
            }
            catch (const invalid_argument&)
            {
              removedSelected = "\""+removedSelected+"\"";
              Obj.removeSong(removedSelected);

            }
            sleep(3);
            break;

          case 4: //exit the sub-menu
            cout<<"Exiting............"<<endl;
            sleep(2);
            break;

          case 99: //exit the program
            cout<<"Exiting............"<<endl;
            sleep(2);
            break;

          default:
            cout<<"Please enter valid input and try again !\n";
            sleep(2);
            break;
        }
        break;
      case 99: //exit the program
        cout<<"Exiting............"<<endl;
        sleep(2);
        break;
      default:
        cout<<"Please enter valid input and try again !\n";
        sleep(2);
        break;
    }
  }
  while(choice != 99 ); //end of main menu loop

  return 0;

}
```
